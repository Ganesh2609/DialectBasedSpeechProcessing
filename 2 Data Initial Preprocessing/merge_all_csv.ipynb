```python
import pandas as pd
import os
```

```python
folder_path = "E:/Work/My Papers/Dravidian Lang Tech 2026/Dialect based speech processing/Dataset/Train/Data CSV"
```

```python
df = pd.concat(
    [
        pd.read_csv(os.path.join(folder_path, f))
        for f in os.listdir(folder_path)
        if f.lower().endswith(".csv")
    ],
    ignore_index=True
)
```

```python
df.to_csv("E:/Work/My Papers/Dravidian Lang Tech 2026/Dialect based speech processing/Dataset/Train/Data CSV/merged.csv", index=False)
```
