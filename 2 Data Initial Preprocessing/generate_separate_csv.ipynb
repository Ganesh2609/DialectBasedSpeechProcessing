```python
import os
import pandas as pd
```

```python
data_path = 'E:/Work/My Papers/Dravidian Lang Tech 2026/Dialect based speech processing/Dataset/Train/Western_Dialect'
df = pd.DataFrame({
    'name': pd.Series(dtype='string'),
    'audio_path': pd.Series(dtype='string'),
    'transcription': pd.Series(dtype='string')
})
```

```python
rows = []

for i in os.listdir(data_path):
    if i.endswith('.txt'):
        file_path = os.path.join(data_path, i)
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                stripped = line.strip()
                name = stripped.split(' ')[0]
                folder_name = '_'.join(line.split('_')[:2]) + '_audio'
                audio_path = os.path.join(data_path, folder_name, name + '.wav')
                transcription = line.split(maxsplit=1)[1].strip()
                rows.append({
                    'name': name,
                    'audio_path': audio_path,
                    'transcription': transcription
                })

df = pd.DataFrame(rows)
```
